```python
import matplotlib.pyplot as plt
import numpy as np
```

```python



labels = ['Urban_nadir', 'Urban_oblique', 'Rural_nadir', 'Rural_oblique']
less_1 = [25,17,5,6]
less_2 = [0,8,11,10]
less_3 = [0,0,7,7]

less_1 = np.array(less_1)/25*100
less_2 = np.array(less_2)/25*100
less_3 = np.array(less_3)/25*100

width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(4,6))

ax1 = ax.bar(labels, less_1, width, label='< 1m',color='#3ba1c5')
ax2 = ax.bar(labels, less_2, width, bottom=less_1,
       label='< 2m',color='#7cc0d8')
ax3 = ax.bar(labels, less_3, width, bottom=less_1+less_2,
       label='< 3m',color='#bee0ec')

for r1, r2, r3 in zip(ax1, ax2, ax3):
    h1 = r1.get_height()
    h2 = r2.get_height()
    h3 = r3.get_height()
    if h1>0:
        plt.text(r1.get_x() + r1.get_width() / 2., h1 , "%d" % h1+ "%", ha="center", va="top", color="black", fontsize=11)
    if h2>0:
        plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 , "%d" % h2+ "%", ha="center", va="top", color="black", fontsize=11)
    if h3>0:
        plt.text(r3.get_x() + r3.get_width() / 2., h1+h2+h3, "%d" % h3+ "%", ha="center", va="top", color="black", fontsize=11)
    

ax.set_ylabel('Percentage of prediction that falls under thresholded error [%]')
ax.set_title('Upsamling by a factor 8')
ax.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)    
plt.xticks(rotation=45, ha='right')
plt.savefig('GSRPT_x8_bar.png',bbox_inches='tight')
plt.show()
```
